```python
null_columns = wine.columns[wine.isnull().any()]
wine[null_columns].isnull().sum()
```

```python
wine['fixed acidity'].fillna(wine['fixed acidity'].mode()[0], inplace=True)
wine['volatile acidity'].fillna(wine['volatile acidity'].mode()[0], inplace=True)
wine['citric acid'].fillna(wine['citric acid'].mode()[0], inplace=True)
wine['residual sugar'].fillna(wine['residual sugar'].mode()[0], inplace=True)
wine['chlorides'].fillna(wine['chlorides'].mode()[0], inplace=True)
wine['free sulfur dioxide'].fillna(wine['free sulfur dioxide'].mode()[0], inplace=True)
wine['total sulfur dioxide'].fillna(wine['total sulfur dioxide'].mode()[0], inplace=True)
wine['density'].fillna(wine['density'].mode()[0], inplace=True)
wine['pH'].fillna(wine['pH'].mode()[0], inplace=True)
wine['sulphates'].fillna(wine['sulphates'].mode()[0], inplace=True)
wine['alcohol'].fillna(wine['alcohol'].mode()[0], inplace=True)
wine['quality'].fillna(wine['quality'].mode()[0], inplace=True)
wine[''].fillna(wine[''].mode()[0], inplace=True)
```

```python
X = wine.drop('quality',1)
y = wine.quality
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=7)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
model_pred = model.predict(X_test)

model_accuracy = round(model.score(X_train,y_train)*100)
print(model_accuracy)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn import model_selection
from sklearn.svm import SVC
model=SVC()
model.fit(X_train, y_train)
model_pred = model.predict(X_test)

model_accuracy = round(model.score(X_train,y_train)*100)
print(model_accuracy)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
from sklearn import model_selection
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model_pred = model.predict(X_test)

model_accuracy = round(model.score(X_train,y_train)*100)
print(model_accuracy)
```
